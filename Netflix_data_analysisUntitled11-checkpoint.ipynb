```python
oldest_release_year = netflix_df["release_year"].min()
newest_release_year = netflix_df["release_year"].max()
print("The oldest content on Netflix was released in " + str(oldest_release_year))
print("The lastest content on Netflix was released in " + str(newest_release_year))
```

```python
netflix_df["year_added"].value_counts().sort_index()
```

```python
netflix_df["year_added"].value_counts().sort_index().plot(kind="line", title="Number of Titles Added per Year", rot=45, color="black")
```

```python
netflix_df[netflix_df["country"] != "Unknown"]["country"].value_counts().head(10).plot(kind="bar", title="Top 10 countries that produce the most content on Netflix", 
                                                   rot=45, color="orange")
```

```python
netflix_df.groupby(["country", "type"]).size().unstack()
```

```python
netflix_df["main_country"] = netflix_df[netflix_df["country"] != "Unknown"]["country"].str.split(",").str[0]
```

```python
netflix_df.groupby(["main_country", "type"]).size().unstack().sort_values(by="Movie", ascending=False).head(10)
```

```python
most_common_movie_duration = netflix_df[netflix_df["type"] == "Movie"]["duration"].mode()[0]
print("Th most common movie duration on Netflix is " + str(most_common_movie_duration) + " mins")
```

```python
movies_more_than_2hrs = netflix_df[(netflix_df["type"] == "Movie") & (netflix_df["duration"] > 120)]
print("Number of Movies longer than 2hrs: " + str(len(movies_more_than_2hrs)))
```

```python
movies_less_than_90mins = netflix_df[(netflix_df["type"] == "Movie") & (netflix_df["duration"] < 90)]
print("Number of short movies on Netflix: " + str(len(movies_less_than_90mins)))
```

```python
netflix_df["listed_in"].head(20)
```

```python
genres = netflix_df["listed_in"].str.split(", ")
genres_exploded = genres.explode()
genres_exploded.value_counts().head(10).plot(kind="bar", title="Top 10 Most Common Genre on Netflix", color="green", rot=45)
```

```python
movie_df = netflix_df[netflix_df["type"] == "Movie"]
movie_genres = movie_df["listed_in"]
movie_genres = movie_genres.str.split(", ")
movie_genres = movie_genres.explode()
movie_genres = movie_genres.value_counts().head(10).plot(kind="bar", title="Top 10 Most Common Genre for Movies on Netflix", color="blue", rot=45)
```

```python
tvshow_df = netflix_df[netflix_df["type"] == "TV Show"]
movie_genres = movie_df["listed_in"]
movie_genres = movie_genres.str.split(", ")
movie_genres = movie_genres.explode()
movie_genres = movie_genres.value_counts().head(10).plot(kind="bar", title="Top 10 Most Common Genre for TV Show on Netflix", color="blue", rot=45)
```

```python
movies_year = movie_df.groupby("release_year")["duration"].median()
movies_year_duration = movies_year.plot(kind="line", xlabel="Released Year", ylabel="Duration")
```

```python
duration_of_movie_before2000 = movie_df[movie_df["release_year"] < 2000]["duration"].median()
duration_of_movie_after2000 = movie_df[movie_df["release_year"] >= 2000]["duration"].median()
print("Older movies are longe than modern movies on Netflix by " + str(duration_of_movie_before2000 - duration_of_movie_after2000) + " mins")
```

```python
movie_len_by_country = movie_df.groupby("main_country")["duration"].median().sort_values(ascending=False)
movie_len_by_country.head(10)
```

```python
counts = movie_df["main_country"].value_counts()
valid_countries = counts[counts > 100].index
filtered_movie = movie_df[movie_df["main_country"].isin(valid_countries)]
filtered_movie.groupby("main_country")["duration"].median().sort_values(ascending=False).head(10)
```

```python
growth = netflix_df.groupby(["year_added", "main_country"]).size()
growth = growth.unstack()
growth = growth.fillna(0)
top_countries = netflix_df["main_country"].value_counts().head(5).index
growth[top_countries].plot(kind="line", figsize=(10,6), title="Growth of Top Countries on Netflix")
```
