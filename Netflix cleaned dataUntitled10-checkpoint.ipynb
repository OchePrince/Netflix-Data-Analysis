```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
netflix_df = pd.read_csv(r"C:\Users\USER\Downloads\netflix_titles.csv")
netflix_df.head()
```

```python
netflix_df.info()
```

```python
netflix_df.isnull().sum()
```

```python
netflix_df.loc[:, "director"] = netflix_df["director"].fillna("Unknown")
netflix_df.loc[:, "cast"] = netflix_df["cast"].fillna("Not Available")
netflix_df.loc[:, "country"] = netflix_df["country"].fillna("Unknown")
netflix_df.loc[:, "rating"] = netflix_df["rating"].fillna("Unknown")
```

```python
netflix_df.isnull().sum()
```

```python
netflix_df.loc[:, "date_added"] = pd.to_datetime(netflix_df["date_added"], format="mixed", errors="coerce")
netflix_df = netflix_df.dropna(subset=["date_added"]) 
```

```python
netflix_df.isnull().sum()
```

```python
print(netflix_df["duration"].dtype)
```

```python
netflix_df["duration"] = netflix_df["duration"].str.extract(r"(\d+)")[0]
netflix_df["duration"] = pd.to_numeric(netflix_df["duration"], errors="coerce")
```

```python
print(netflix_df["duration"].dtype)
```

```python
netflix_df["duration"] = netflix_df["duration"].fillna(netflix_df["duration"].median())
netflix_df["duration"] = netflix_df["duration"].astype(int)
```

```python
netflix_df.isnull().sum()
```

```python
netflix_df["date_added"] = pd.to_datetime(netflix_df["date_added"], errors="coerce")
netflix_df["month_added"] = netflix_df["date_added"].dt.month
netflix_df["year_added"] = netflix_df["date_added"].dt.year
```

```python
netflix_df.head()
```

```python
netflix_df.loc[:, "director"] = netflix_df["director"].str.strip()
netflix_df.loc[:, "cast"] = netflix_df["cast"].str.strip()
netflix_df.loc[:, "country"] = netflix_df["country"].str.strip()
netflix_df.loc[:, "description"] = netflix_df["description"].str.strip()
netflix_df.loc[:, "type"] = netflix_df["type"].str.strip()
netflix_df.loc[:, "listed_in"] = netflix_df["listed_in"].str.strip()
netflix_df.loc[:, "rating"] = netflix_df["rating"].str.strip()
```
